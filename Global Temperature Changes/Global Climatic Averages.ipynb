```python
# Adam Benson
# Climate and Space Sciences Engineering 405-002
# Lab Report One
# 1/10/2018
# Import Statements
import numpy as np #for scientific calculations
import pandas as pd #for data analysis
import matplotlib.pyplot as plt #for visualization
%matplotlib inline
```

### Loading Land Data

```python
 #Creating dataframe from land data
globalLandTemp = pd.read_csv('./Data/GlobalLand_1880-2017_TempVariants.csv',
                              header = 4)
# Specifying that value is Temp in C
globalLandTemp.rename(columns={'Value': 'AnomalyC'}, inplace=True) 
# Useful for US audience - creating column: Anomaly Temperature in Fahrenheit 
globalLandTemp['AnomalyF'] = globalLandTemp['AnomalyC'] * 1.8 
```

```python
# Checking to make sure that data was loaded correctly
print('HEAD\n', globalLandTemp.tail(n=3), '\n') 
print('TAIL\n', globalLandTemp.head(n=3))
```

### Loading Ocean Data

```python
 #Creating dataframe from ocean data
globalOceanTemp = pd.read_csv('./Data/GlobalOcean_1880-2017_TempVariants.csv',
                               header = 4)
# Specifying that value is Temp in C
globalOceanTemp.rename(columns={'Value': 'AnomalyC'}, inplace=True)
# Useful for US audience - creating Temp in F column
globalOceanTemp['AnomalyF'] = globalOceanTemp['AnomalyC'] * 1.8


warmTemperaturesMask = globalOceanTemp['AnomalyF'] > 0 
#print(type(booleanIndex))
#print(booleanIndex[:10])

print('HEAD\n', globalOceanTemp.tail(n=3), '\n')
print('TAIL\n', globalOceanTemp.head(n=3))                              
```

## Visualization Component

### Line Charts

```python
# Creating a new column 'Colors' in both the land and ocean dataframes
# This column will be used in the visualization
# A mask will later be applied to change the colors to reveal trends

globalLandTemp['Colors'] = 'g'
globalOceanTemp['Colors'] = 'b'
```

```python
landLinePlot = plt.figure(figsize=(12, 5))
landLinePlot.suptitle('Global Land Temperature Anomalies, 1880-2016', fontsize=20)
plt.plot(globalLandTemp['Year'], globalLandTemp['AnomalyC'], color='g')

# Dashed Horizontal line at y=0 to easily visualize positive and negative anomalies
plt.axhline(y=0.0, color='k', linestyle='dashed')

## Ticks and Labels
plt.xlabel("Year", fontsize=18)
plt.xticks(fontsize=14)
plt.ylabel("Anomaly", fontsize=18)
plt.yticks(fontsize=14)
plt.show() # show chart
```

```python
# Specifying figure dimensions
oceanLinePlot = plt.figure(figsize=(12, 5))
oceanLinePlot.suptitle('Global Ocean Temperature Anomalies, 1880-2016', fontsize=20)
plt.plot(globalOceanTemp['Year'], globalOceanTemp['AnomalyC'], color='b')


# Dashed Horizontal line at y=0 to easily visualize positive and negative anomalies
plt.axhline(y=0.0, color='k', linestyle='dashed')

## Ticks and Labels
plt.xlabel("Year", fontsize=18)
plt.xticks(fontsize=14)
plt.ylabel("Anomaly", fontsize=18)
plt.yticks(fontsize=14)

plt.show() # show chart
```

### Bar Charts

```python
# Plotting bar charts to better display temperature anomalies
# Temp Anomaly (in Fahrenheit) compared to the global mean temp is the Y axis
# Time Series so, Year = X axis
```

```python
# Creating masks to change colors based on severity of AnomalyF
temperatureMask = globalOceanTemp['AnomalyF'] > 0 # positive AnomalyF (temperature increase)
highTempMask = globalOceanTemp['AnomalyF'] > 1.0 # Very warm year

# Where True, change color to yellow (warming)
globalOceanTemp.loc[temperatureMask, 'Colors'] = 'y' 
# Where False, change color to cyan (cooling)
globalOceanTemp.loc[~temperatureMask, 'Colors'] = 'c' 
# Where True, representing very high anomaly, change to red
globalOceanTemp.loc[highTempMask, 'Colors'] = 'r' 



oceanBarChart = plt.figure(figsize=(10, 5))
oceanBarChart.suptitle('Ocean Temperature Variants, 1800 - 2017 \n (Global Averages)', fontsize=20)


# align = 'edge' will align to the left side of the range
# width = 0.8 rather than 1 simply to make it appear more interesting. 

plt.bar(globalOceanTemp['Year'], globalOceanTemp['AnomalyF'], width = 0.8, align='edge', 
        color = globalOceanTemp['Colors'])
#----------------------------------------- 
plt.ylabel('Anomaly $\degree$F', fontsize = 20) # Y-Axis = Temp(F)
plt.xlabel('Year', fontsize = 20) # X-Axis = Year

plt.xticks(fontsize=16) # increase size
plt.yticks(fontsize=16) # increase size
# Horizontal line at y=0
# Because values can be both positive and negative
plt.axhline(y=0.0, color='k', linestyle='--') 
                                              
plt.show() # show chart
```

```python
# Creating masks to change colors based on severity of AnomalyF

# positive AnomalyF (temperature increase)
warmTempMask = globalLandTemp['AnomalyF'] > 0 
# Very warm year
veryWarmTempMask = globalLandTemp['AnomalyF'] > 1.0 

# Where True, change col or to yellow (warming)
globalLandTemp.loc[warmTempMask, 'Colors'] = 'y' 
# Where False, change color to cyan (cooling)
globalLandTemp.loc[~warmTempMask, 'Colors'] = 'c' 
# Where True, change color to red (very high anomaly)
globalLandTemp.loc[veryWarmTempMask, 'Colors'] = 'r' 

# Setting up figure
landBarChart = plt.figure(figsize=(10, 5))
landBarChart.suptitle('Global Land Temperature Anomalies, 1880-2016 \n (Global Averages) ', fontsize=20)


# align = 'edge' will align to the left side of the range
# width = 0.8 rather than 1 simply to make it appear more interesting. 
plt.bar(globalLandTemp['Year'], globalLandTemp['AnomalyF'], width = 0.8, align='edge', 
        color = globalLandTemp['Colors'])
#----------------------------------------- 
plt.ylabel('Anomaly $\degree$F', fontsize = 20) # Y-Axis = Temp(F)
plt.xlabel('Year', fontsize = 20) # X-Axis = Year

plt.xticks(fontsize=16) # increase size
plt.yticks(fontsize=16) # increase size
# Horizontal line at y=0
# Because values can be both positive and negative
plt.axhline(y=0.0, color='k', linestyle='dashed') 
plt.show()
```

# Analysis

Global ocean temperatures are rising slower than global land temperatures. Both land and ocean temperatures, however, demonstrate positive trends over the past 40 years. One explanation for the differences in global temperatures based on the type of surface is the fact that water has a heat capacity of 4.18 J/gC while soil has a heat capacity of 0.8 J/gC. This means that temperature changes occur faster on land masses than they do in bodies of water. Thus, it makes sense that global land temperatures are rising faster than global ocean temperatures. Lastly, it is worth noting that specific locations can experience a wide variety of temperatures compared to the average. This means that some areas may be experiencing trends not represented in the charts above. Regardless, 2016 is the warmest year on record since 1880. 2017 is the second warmest. Ultimately, there is clear evidence that our planet is warming.
